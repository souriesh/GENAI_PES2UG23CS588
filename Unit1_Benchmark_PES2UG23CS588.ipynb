```python
from transformers import pipeline
```

# **TEXT GENERATION**

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}

for name, model in models.items():
    try:
        generator = pipeline("text-generation", model=model)
        output = generator("Artificial Intelligence will change the world because", max_length=40)
        print(f"\n{name} Output:\n", output)
    except Exception as e:
        print(f"\n{name} Error:\n", e)
```

# FILL **MASK**

```python
for name, model in models.items():
    try:
        masker = pipeline("fill-mask", model=model)
        output = masker("Artificial Intelligence is <mask> the future of technology.")
        print(f"\n{name} Output:\n", output[:3])
    except Exception as e:
        print(f"\n{name} Error:\n", e)
```

##QUESTION ANSWERING

```python
for name, model in models.items():
    try:
        qa = pipeline("question-answering", model=model)
        output = qa({
            "context": "Artificial Intelligence is a branch of computer science that focuses on building systems capable of performing tasks that normally require human intelligence.",
            "question": "What does Artificial Intelligence focus on?"
        })
        print(f"\n{name} Output:\n", output)
    except Exception as e:
        print(f"\n{name} Error:\n", e)
```

## Conclusion
- Encoder-only models (BERT, RoBERTa) excel at understanding tasks
- Encoder–Decoder models (BART) excel at generation
- Model architecture determines task suitability
